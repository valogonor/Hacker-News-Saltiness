```python
import pandas as pd
import re
from textblob import TextBlob
hn1 = pd.read_csv('hackernews1.csv')
hn1.head()
```

```python
def get_sentiment(text):
    """
    Utility function to classify sentiment of passed text
    using textblob's sentiment method. Return the polarity
    score as a float within the range [-1.0, 1.0]
    """
    return TextBlob(text).sentiment.polarity
```

```python
hn1.shape
```

```python
hn1 = hn1.dropna()
hn1.shape
```

```python
hn1['sentiment'] = hn1.text.apply(lambda x: get_sentiment(x))
hn1 = hn1.sort_values(by=['sentiment'], ascending=True)
hn1.head()
```

```python
hn1.to_csv('hackernews1.csv')
```
